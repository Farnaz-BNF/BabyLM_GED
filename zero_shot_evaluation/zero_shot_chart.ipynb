```python
import os
import json
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def read_model_scores(path):
  scores = {}
  path = os.path.join(path, "zeroshot_with_prep")
  for dir_name in os.listdir(path):
    dir_path = os.path.join(path, dir_name)
    if os.path.isdir(dir_path):
      result_path = os.path.join(dir_path, "eval_results.json")
      with open (result_path, "r") as result_file:
        result = json.load(result_file)
        # print(f"{dir_name}: {result['eval_accuracy']}")
        scores[dir_name] = result['eval_accuracy']
  return scores
```

```python
def read_all_models_scores():
  all_scores = {}
  for i in range(1, 11):
    model_path = f"/content/drive/MyDrive/VU Thesis/Code/baby_models/V20L4/model_e{i}_v20_l4"
    all_scores[f"model_e{i}_v20_l4"] = read_model_scores(model_path)
  return all_scores
```

```python
scores = read_all_models_scores()
scores
```

```python
scores_df = pd.DataFrame(scores)
# scores_df.T.head(15)
scores_df.head(15)
```

```python
means = scores_df.mean()
scores_df.loc["mean"] = means
scores_df.head(15)
```

```python
scores_df = scores_df.T
scores_df.head(15)
```

```python
scores_df.drop(["anaphor_agreement",	"argument_structure",	"binding",	"control_raising",	"ellipsis",	"filler_gap",	"irregular_forms",	"island_effects",	"npi_licensing",	"quantifiers"], axis= 1, inplace=True)
scores_df.head(15)
```

```python
scores_df.index = ["E1", "E2", "E3", "E4", "E5", "E6", "E7", "E8", "E9", "E10"]
scores_df.head(15)
```

```python
plt.figure(figsize= (10,6))
for col in scores_df.columns:
  plt.plot(scores_df.index, scores_df[col], marker = "o", label= col)
plt.legend(loc="upper right", bbox_to_anchor = (1.35, 1))
plt.show()
```

```python
roberta_score = {}
roberta_score["roberta_base"] = read_model_scores("/content/drive/MyDrive/VU Thesis/Code/baby_models/roberta-base")
roberta_score
```

```python
roberta_scores_df = pd.DataFrame(roberta_score)
mean = roberta_scores_df.mean()
roberta_scores_df.loc["mean"] = mean
roberta_scores_df = roberta_scores_df.T
roberta_scores_df.drop(["anaphor_agreement",	"argument_structure",	"binding",	"control_raising",	"ellipsis",	"filler_gap",	"irregular_forms",	"island_effects",	"npi_licensing",	"quantifiers"], axis= 1, inplace=True)

roberta_scores_df.head(15)
```

```python
import matplotlib.pyplot as plt

# Define consistent color mapping for unique error types
colors = {
    "determiner_noun_agreement": "tab:blue",
    "subject_verb_agreement": "tab:red",
    "preposition": "tab:green",
    "mean": "tab:orange"
}

plt.figure(figsize=(10, 6))

# Plot each error type once with the same color across both dataframes
for col in scores_df.columns:
    color = colors.get(col, None)
    plt.plot(scores_df.index, scores_df[col], marker="o", label=col, color=color)
    plt.plot(roberta_scores_df.index, roberta_scores_df[col], marker=".", color=color)

plt.legend(loc="upper right", bbox_to_anchor=(1.35, 1))
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define consistent color mapping for unique error types
colors = {
    "determiner_noun_agreement": "tab:orange",
    "subject_verb_agreement": "tab:green",
    "preposition": "tab:red",
    "mean": "tab:blue"
}

plt.figure(figsize=(10, 6))
ax = plt.gca()

# Plot each error type once with same color and show values
for col in scores_df.columns:
    color = colors.get(col, None)

    # Solid markers for scores_df
    ax.plot(scores_df.index, scores_df[col], marker="o", label=col, color=color)

    # Add values for scores_df
    for i, (x, y) in enumerate(zip(scores_df.index, scores_df[col])):
        ax.text(x, y + 0.01, f"{y:.2f}", ha='center', va='bottom', fontsize=8, color=color)

    # Hollow markers for roberta_scores_df
    ax.plot(roberta_scores_df.index, roberta_scores_df[col], marker="o", linestyle='-',
            markerfacecolor='none', markeredgecolor=color, color=color)

    # Add values for roberta_scores_df
    for i, (x, y) in enumerate(zip(roberta_scores_df.index, roberta_scores_df[col])):
        ax.text(x, y - 0.03, f"{y:.2f}", ha='center', va='top', fontsize=8, color=color)

plt.legend(loc="upper right", bbox_to_anchor=(1.35, 1))
plt.show()
```
